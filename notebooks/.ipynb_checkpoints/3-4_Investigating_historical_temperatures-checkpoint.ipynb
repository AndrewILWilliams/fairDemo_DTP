```python
#Read in observations, and remove 1861-80 climatology
Y_obs, T_obs = np.genfromtxt("../HadCRUT.4.5.0.0.annual_ns_avg.txt")[:,0], np.genfromtxt("../HadCRUT.4.5.0.0.annual_ns_avg.txt")[:,1]
i_obs_clm=(1861<=Y_obs) & (Y_obs<=1880)
T_obs=T_obs-np.mean(T_obs[i_obs_clm])
```

## Simulating the 20th century temperature record with `FaIR`

```python
"""Set key model parameters"""
ECS_def = 2.7     # equilibrium climate sensitivity (K), default=2.7K
TCR_def = 1.6     # transient climate response (K), default=1.6K
tau_s_def = 4.1   # ocean mixed layer response time scale (years), default=4.1yrs
tau_d_def = 239.0 # deep ocean response time scale (years), default=239.0yrs


"""Set scalings for the anthropogenic GHG and aerosol and natural forcings (defaults=1.0)"""
ghg_scale=1.0 
aer_scale=1.0 
nat_scale=1.0 

""" Read in radiative forcing to 2020 """
years,F_tot,F_ant,F_ghg,F_co2=load_rcp_forcing('../RCP85_MIDYEAR_RADFORCING.csv',Y_max=2020)

#Get aerosol forcing as difference between total anthropogenic forcing and that due to greenhouse gases. 
F_aer=F_ant-F_ghg

#Get natural forcing as difference between total forcing and total anthropogenic forcing
F_nat=F_tot-F_ant

#Scale the forcings
F_ghg=F_ghg*ghg_scale
F_co2=F_co2*ghg_scale
F_aer=F_aer*aer_scale
F_nat=F_nat*nat_scale

#Get new total and anthropogenic forcings
F_ant=F_ghg+F_aer
F_tot=F_ant+F_nat


# Run the model with full forcing and decomposed forcing 
_,F_tot,T_tot = fair_scm(
    emissions=False,                     # Just drive the model using radiative forcing
    other_rf = F_tot,                  # Input time-series of radiative forcing
    useMultigas=False,                   
    tcrecs=np.array([TCR_def, ECS_def]),
    d=np.array([tau_d_def, tau_s_def]))


_,F_ant,T_ant = fair_scm(
    emissions=False,                     # Just drive the model using radiative forcing
    other_rf = F_ant,                  # Input time-series of radiative forcing
    useMultigas=False,                   
    tcrecs=np.array([TCR_def, ECS_def]),
    d=np.array([tau_d_def, tau_s_def]))


_,F_ghg,T_ghg = fair_scm(
    emissions=False,                     # Just drive the model using radiative forcing
    other_rf = F_ghg,                  # Input time-series of radiative forcing
    useMultigas=False,                   
    tcrecs=np.array([TCR_def, ECS_def]),
    d=np.array([tau_d_def, tau_s_def]))

_,F_aer,T_aer = fair_scm(
    emissions=False,                     # Just drive the model using radiative forcing
    other_rf = F_aer,                  # Input time-series of radiative forcing
    useMultigas=False,                   
    tcrecs=np.array([TCR_def, ECS_def]),
    d=np.array([tau_d_def, tau_s_def]))



#Remove 1861-80 climatology
i_clm=(1861<=years) & (years<=1880)
T_tot=T_tot-np.mean(T_tot[i_clm])
T_ant=T_ant-np.mean(T_ant[i_clm])
T_ghg=T_ghg-np.mean(T_ghg[i_clm])
T_aer=T_aer-np.mean(T_aer[i_clm])


print("Simulated anthropogenic warming in 2015:",round(T_ant[years==2015][0],2))

#Compute standard deviation of differences of simulated temperature from observations (a measure of the 
#goodness of fit, with smaller values implying a better fit)
T_std=np.std(T_obs[(1861<=Y_obs) & (Y_obs<=2016)] - T_tot[(1861<=years) & (years<=2016)])
print("Standard deviation of error of total response (K)",round(T_std,3))

#Now plot results
plt.figure(dpi=100, figsize = (8,8))
plt.subplot(211)
plt.axis([1860., 2020., -1.5, 3.5])
plt.plot(years,F_tot,color='red',linewidth=2)
plt.plot(years,F_ant,color='orange')
plt.plot(years,F_ghg,color='magenta')
plt.plot(years,F_aer,color='blue')
plt.plot(years,F_nat,color='green')
plt.ylabel("Forcing (W m$^{-2}$)")
plt.xlabel("Year")

plt.subplot(212)
plt.axis([1860., 2020., -0.5, 1.5])
plt.plot(Y_obs,T_obs,color='black',marker='.',ls='None',label='obs')
plt.plot(years,T_tot,color='red',label='Total',linewidth=2)
plt.plot(years,T_ant,color='orange',label='Total anthro.')
plt.plot(years,T_ghg,color='magenta',label='GHGs')
plt.plot(years,T_aer,color='blue',label='Aerosols')
plt.plot(years,T_tot-T_ant,color='green',label='Nat.')
plt.ylabel("Warming ($^\circ$C)")
plt.xlabel("Year")
plt.legend(loc='upper left')
```
